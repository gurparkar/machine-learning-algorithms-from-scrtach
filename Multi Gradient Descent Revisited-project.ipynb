```python
# m is clearly a matrix of (n+1)*1 order because we add another 
#feature x[i]=1 in data[m*n+1]

def cost(data,Y,m):
    Cost=0
    M=len(data)
    Cost=(1/M)*sum((data.dot(m)-Y.reshape(379,1))**2)
    return Cost
```

```python
def step_gradient(data,Y,lRate,m):
    M=len(data)
    N=data.shape[1]
    sum_=np.zeros((N,1)) 
    m=m-(-2/M)*lRate*(data.T).dot(Y.reshape(379,1)-data.dot(m))   
    #print m.shape
    return m

def gradient_descent(data,Y,lRate,numIter):
    #there is no c here
    N=data.shape[1]
    m=np.zeros((N,1))
    
    for i in range(numIter):        
        m=step_gradient(data,Y,lRate,m) 
    return m
```
